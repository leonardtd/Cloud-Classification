```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="whitegrid")
```

---

```python
path = "data_entries/RE5 resultados.csv"
```

```python
df = pd.read_csv(path)
```

```python
df.shape
```

```python
drop_cols = [c for c in df.columns if ("MIN" in c) or ('MAX' in c) or ('Step' in c)]
df.drop(columns=drop_cols, inplace=True)
df['epoch'] = (df.index + 1)
```

```python
df.columns.tolist()
```

---

# 1. Cosine Similarity

### Adjacency density

```python
df['cos_sim 0.85 - test/test_adj_density'].mean()
```

```python
df['cos_sim 0.7 - test/test_adj_density'].mean()
```

```python
df['cos_sim 0.6 - test/test_adj_density'].mean()
```

```python
df['cos_sim 0.6 - test/test_adj_density'][0]
```

```python
fig, ax = plt.subplots(1,1, dpi=1000)
fig.set_figwidth(8)
fig.set_figheight(5)

ax.title.set_text(r"Densidad de la matriz de adyacencia por valor del umbral $\tau$")

ax.plot(df['epoch'], df['cos_sim 0.85 - test/test_adj_density']*100, label= r'$\tau=0.85$')
ax.plot(df['epoch'], df['cos_sim 0.7 - test/test_adj_density']*100, label= r'$\tau=0.70$')
ax.plot(df['epoch'], df['cos_sim 0.6 - test/test_adj_density']*100, label= r'$\tau=0.60$')

ax.set_xlabel('Epoch', fontsize=9)
ax.set_ylabel('Density (%)', fontsize=10)

plt.legend()

fig.tight_layout()
fig.savefig('assets/adj_cos_sim_re2_2.png')
plt.show()
```

### Accuracy

```python
df['cos_sim 0.85 - test/test_accuracy'].max()
```

```python
df['cos_sim 0.7 - test/test_accuracy'].max()
```

```python
df['cos_sim 0.6 - test/test_accuracy'].max()
```

```python
fig, axs = plt.subplots(1,2, dpi=1000)
fig.set_figwidth(14)
fig.set_figheight(5)
fig.suptitle(r"Mtricas de evaluacin por valor del umbral $\tau$", fontsize=14)


axs[0].title.set_text(r"Precisin")

axs[0].plot(df['epoch'], df['cos_sim 0.85 - test/test_accuracy']*100, label= r'$\tau=0.85$')
axs[0].plot(df['epoch'], df['cos_sim 0.7 - test/test_accuracy']*100, label= r'$\tau=0.70$')
axs[0].plot(df['epoch'], df['cos_sim 0.6 - test/test_accuracy']*100, label= r'$\tau=0.60$')

axs[0].set_xlabel('Epoch', fontsize=9)
axs[0].set_ylabel('Accuracy (%)', fontsize=10)
axs[0].legend()


axs[1].title.set_text(r"Funcin de Prdida")

axs[1].plot(df['epoch'], df['cos_sim 0.85 - test/test_loss'], label= r'$\tau=0.85$')
axs[1].plot(df['epoch'], df['cos_sim 0.7 - test/test_loss'], label= r'$\tau=0.70$')
axs[1].plot(df['epoch'], df['cos_sim 0.6 - test/test_loss'], label= r'$\tau=0.60$')

axs[1].set_xlabel('Epoch', fontsize=9)
axs[1].set_ylabel('Cross Entropy Loss', fontsize=10)
axs[1].legend()


plt.subplots_adjust(left=0.05,
                    bottom=0.1, 
                    right=1-0.01, 
                    top=0.85, 
                    wspace=0.2, 
                    hspace=0)

#fig.tight_layout()
fig.savefig('assets/metrics_cos_sim_re2_2.png')
plt.show()
```

### Loss

```python
df['cos_sim 0.6 - test/test_loss'].min()
```

# 3. L2 distance

### Adjacency

```python
fig, ax = plt.subplots(1,1, dpi=1000)
fig.set_figwidth(8)
fig.set_figheight(5)

ax.title.set_text(r"Densidad de la matriz de adyacencia por valor $\bf k$")

ax.plot(df['epoch'], df['l2 2 - test/test_adj_density']*100, label= r'$k=2$')
ax.plot(df['epoch'], df['l2 6 - test/test_adj_density']*100, label= r'$k=6$')
ax.plot(df['epoch'], df['l2 15 - test/test_adj_density']*100, label= r'$k=15$')

ax.set_xlabel('Epoch', fontsize=9)
ax.set_ylabel('Density (%)', fontsize=10)

plt.legend()

fig.tight_layout()
fig.savefig('assets/adj_l2_distance_re2_2.png')
plt.show()
```
