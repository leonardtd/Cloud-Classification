```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

---

```python
path = "data_entries/cos_sim_0_6.csv"
```

```python
data = pd.read_csv(path)
```

```python
drop_cols = [c for c in data.columns if ("MIN" in c) or ('MAX' in c) or ('Step' in c)]
```

```python
data.drop(columns=drop_cols, inplace=True)
```

```python
data['epoch'] = (data.index + 1)
```

---

```python
columns = data.columns
columns = [c.split('/')[-1] for c in columns]
```

```python
data.columns = columns
```

---

### plot

```python
columns
```

```python
sns.set_theme(style="whitegrid")
%config InlineBackend.figure_format = 'svg'
```

```python
fig, axs = plt.subplots(1,3, dpi=1200)
fig.suptitle("Métricas del modelo base por tipo de conjunto de datos", fontsize=16)
fig.set_figwidth(12)
fig.set_figheight(4)


### ACC
axs[0].plot(data['epoch'], data['train_accuracy']*100, label='Entrenamiento')
axs[0].plot(data['epoch'], data['test_accuracy']*100, label='Prueba')
axs[0].title.set_text('Precisión')
axs[0].title.set_fontsize(12)
axs[0].set_xlabel('Epoch', fontsize=9)
axs[0].set_ylabel('Accuracy (%)', fontsize=10)
#axs[0].legend()


### LOSS
axs[1].plot(data['epoch'], data['train_loss'], label='Entrenamiento')
axs[1].plot(data['epoch'], data['test_loss'], label='Prueba')
axs[1].title.set_text('Función de pérdida')
axs[1].title.set_fontsize(12)
axs[1].set_xlabel('Epoch', fontsize=9)
axs[1].set_ylabel('Cross Entropy Loss', fontsize=10)
#axs[1].legend()

### DENSITY
axs[2].plot(data['epoch'], data['train_adj_density']*100, label='Entrenamiento')
axs[2].plot(data['epoch'], data['test_adj_density']*100, label='Prueba')
axs[2].title.set_text('Densidad de la matriz de adyacencia')
axs[2].title.set_fontsize(12)
axs[2].set_xlabel('Epoch', fontsize=9)
axs[2].set_ylabel('Density (%)', fontsize=10)
#axs[2].legend()

lines_labels = [ax.get_legend_handles_labels() for ax in fig.axes][:1] #only first legend
lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]
fig.legend(lines, labels, bbox_to_anchor=(0.5, -0.00), mode='expand', ncol=2, frameon=False)

plt.subplots_adjust(left=0,
                    bottom=0, 
                    right=0.8, 
                    top=0.7, 
                    wspace=0.4, 
                    hspace=0)

fig.tight_layout()

fig.savefig('assets/metricas_re1_3.png')
plt.show()
```
